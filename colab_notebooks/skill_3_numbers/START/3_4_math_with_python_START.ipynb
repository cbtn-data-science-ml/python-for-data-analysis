## Order of Operations

PEMDAS


1.   **P**arentheses `()`
2.   **E**xponent `**`
3.   **M**ultiplication `*`
4.   **D**ivision `\`
5.   **A**ddition `+`
6.   **S**ubstraction `-`


```python
# what will the following code output? 
print(1 + 2 * 4) 
print((1 + 2) * 4)
```

How about sqauring numbers? 

For example: $\large{2^2}$ 

```python
# what will the following code output? 
print(2**2)
print(2**2 / 2)
print(2**2 / 2 + 2)
print(2**2 / (2 + 2))
```
