# String methods
String methods are built-in functions that work on strings. If you remember the definition of methods, they are functions attched to certain objectsâ€”in this case functions attached to strings.

- **capitalize()** - converts the first character to upper case
- **upper()** - converts characters to upper case
- **lower()** - converts characters to lower case
- **split()** - splits words into a comma separated list




### capitalize()

```python
# store "first word in a sentence" inside the variable: my_caps
my_caps = "first word in a sentence"

# capitalize the first word in a sentence
print(my_caps.capitalize())
```

### upper()



```python
# store "all upper case" inside the variable: make_upper
make_upper = "all upper case"

# convert make_upper into all upper caps
print(make_upper.upper())
```

### lower()

```python
# store "all upper case" inside the variable: my_string
my_string = "ALL LOWER CASE"

# convert make_upper into all lower caps
print(my_string.lower())
```

### split()

```python
# store "all upper case" inside the variable: my_string
song = "all the things you are"

# split each word into a comma separated list (defaul is whitespace)
split_song = song.split()

# print split_song
print(split_song)
```

### Variable reassignment
Variable reassignment will come in handy as a data analyst. For example, when working with Python and data, you will often create new versions of a table or dataset and variable reassignment is a popular method of accomplishing this task.

```python
# this makes the split operation on song "permanent"
song = song.split()

# print song
print(song)
```
