# Indexing strings

You can index strings to grab one character from a sequence, or string, of characters. To do this, we will need to use something called a 0 based-index; meaning, we start counting from 0. For example, the first character of the string sequence, `"Jonathan"`, is `J`, and its index is 0. The second letter of the string sequence, `"Jonathan"`, is `o`, its index is 1, and so on.


## Index visualization:
J | O | N | A | T | H | A | N 
--- | --- | --- | --- | --- | --- | --- | --- 
0 | 1 | 2 | 3 | 4 | 5 | 6 | 7 



```python
# indexing a string's first element
"Jonathan"[0]
```

```python
# indexing a string's second element
"Jonathan"[1]
```

```python
# indexing a string's first element
"Jonathan"[7]
```

# Indexing variables
You can also index varibles if they are assigned to strings. There is no difference.

```python
# store "Jonathan" inside variable: name
name = "Jonathan"

# access the first element of: name
name[0]
```

## Negative index visualization:
J | O | N | A | T | H | A | N 
--- | --- | --- | --- | --- | --- | --- | --- 
-8 | -7 | -6 | -5 | -4 | -3 | -2 | -1 

```python
# access the last element of: name
name[-1]
```

```python
# access the penultimate element of: name
name[-2]
```

## Slicing strings

In addition to indexing, you can specify a start, stop and step parameter:


The third parameter is optional, for setting a step:



```python
# slicing using a variable: name
name = "Jonathan"

# slice 0-3, not including 3
name[0:3]
```

The third parameter is optional, for setting a step:


```python
# slicing using a variable: name
name = "Jonathan"

# slice 0-8, not including 8, in steps of 2 (every other element)
name[0:8:2]
```

To reverse a string, don't specify the start and stop paramenters, and enter `-1` to start at the end and stop at the beginning:


```python
# reversing the variable: name
name = "Jonathan"

# this is one way
name[::-1]
```
