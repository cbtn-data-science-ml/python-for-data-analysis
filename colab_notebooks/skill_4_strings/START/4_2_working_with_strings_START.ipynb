## String Formatting

The placeholders can include named indexes `{name}`, numbered indexes `{0}`, and empty placeholders `{}`. Here are three examples:

```python

example_2 = "My name is {0}".format("John Coltrane")
example_3 = "My name is {}".format("John Coltrane")
```
