## Function Scope

Function scope affects the code inside of a function, which is straightforward. However, scope can be tricky when writing complex functions in a large codebase. So what can you do? Exploring the 4 types of scope will help you navigate LEGB like a pro!

To help you remember the four scopes in Python, use the `LEGB` name lookup procedure. The `LEGB` acronym is the order in which names are looked up in Python. Here are some definitions to go with the acronym.

1. `L`ocal scope ¬ª the scope inside of a function
1. `E`nclosing function scope ¬ª the scope in a nested function
1. `G`lobal scope ¬ª the top level scope, for modules‚Äîno indentation
1. `B`uilt-in scope ¬ª preassigned scope, for built-in functions

### Local scope

If you indent your code, you are in in local scope and any variables and functions in local scope will not be available in global scope. For example, if you call your function in the local scope, Python will not execute the function.

Let's write a function that squares a number to observe this behavior. In it's simplest form, you could write:

```python
# ask user to input a value that will get assigned to the variable number
number = int(input("Enter an integer from 0 to 9: "))

# define a function named squared and pass in num as the argument
def squared(num):
  # return num squared
  return num ** 2

  squared(number) # this function call is in local scope
```

If you define a variable in local scope, it will not be available in global scope because local scope isn‚Äôt visible outside of the function where it is created.

```python
# ask user to input a value that will get assigned to the variable number
number = int(input("Enter an integer from 0 to 9: "))

# define a function named squared and pass in num as the argument
def squared(num):
  squared_num = num ** 2 # local scope
  return squared_num 

squared(number) # here we call the function

print(squared_num) # squared_num is in local scope while print is in global scope
```

### Enclosing scope
Enclosing scope only exists inside nested functions(functions defined inside another function). Enclosing scope is only visible inside of inner and enclosing functions.

```python
# input() will ask you to input an value that will get assigned to the variable number
number = int(input("Enter a positive integer: "))

# define a function named squared and pass in num as the parameter
def squared(num):
  squared_num = num ** 2
  def message():
    text = str(num) + " squared is " + str(squared_num) # squared_num is available in enclosing scope
    print(text)
  message()
squared(number)

# print(text) is not available in global scope
```

### Global scope
When you start a Python program, you are in global scope. Any function names or variables you create in this scope will be available globally‚Äîinside or outside of a function.

```python
# ask user to input a value that will get assigned to the variable number
number = int(input("Enter an integer from 0 to 9: "))

# define a function named squared 
def squared():
  # return num squared
  return number ** 2 # number is available since it is in global scope

squared()
```

### Built-in scope

Built-in scope is available from anywhere in your code because keywords, built-in functions, and other attributes are baked into Python. For example, you can call built-in functions like `len()` inside of a function because built-in function scope is preassigned. 

```python
# input() will ask you to input an value that will get assigned to the variable number
number = input("Enter an integer from 0 to 9: ")

# define a function named squared and pass in num as the argument
def squared(num):
  if len(num) > 1:
    print("Enter a number between 0 - 9. Game over.")
  else:
    squared_numss = int(num) ** 2
    print(squared_numss)

squared(number)
```
